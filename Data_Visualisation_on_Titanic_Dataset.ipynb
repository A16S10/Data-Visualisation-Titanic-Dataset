```python
# Write your code here
# Activity 3
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

titanic_dataset=sns.load_dataset('titanic')
titanic_dataset

mean_age_male=titanic_dataset[titanic_dataset['sex']=='male']['age'].mean().round(0)
mean_age_female=titanic_dataset[titanic_dataset['sex']=='female']['age'].mean().round()

print(mean_age_male)
print(mean_age_female)

# fill NaN values with corresponding mean values


titanic_dataset.loc[titanic_dataset['sex']=='male','age']=titanic_dataset.loc[titanic_dataset['sex']=='male','age'].fillna(mean_age_male)

titanic_dataset.loc[titanic_dataset['sex']=='female','age']=titanic_dataset.loc[titanic_dataset['sex']=='female','age'].fillna(mean_age_female)



#Reaplacing age of child below age 1 to 1
titanic_dataset[titanic_dataset['age']<1]

#Converting datatype from int to float
titanic_dataset['survived']=titanic_dataset['survived'].astype('float')

# Scatter plot



# Calculate the IQR
Q1 = titanic_dataset["age"].quantile(0.25)
Q3 = titanic_dataset["age"].quantile(0.75)
IQR = Q3 - Q1

# Define the upper and lower bounds for outliers
upper_bound = Q3 + 1.5 * IQR
lower_bound = Q1 - 1.5 * IQR

# Create a boxplot to show the distribution of age
sns.boxplot(data=titanic_dataset, x="age")



# Show the plot
plt.show()


#  Use Seaborn to create appropriate visualizations such as bar charts, count plots, or heatmaps to explore the relationship between age, gender, and survival status of the passengers on the Titanic.

sns.scatterplot(data=titanic_dataset, x="age", y="survived", hue="who")
plt.show()
###conclusion: Most of the men died in accident.Most women susrvived.

# comparison between survival of men and women
sns.countplot(data=titanic_dataset,x="survived", hue="who")
plt.show()

# Creating heatmap to check relationship between age , gender and survival.

titanic_dataset['Gender_num']=titanic_dataset['who'].map({'man':0,"woman":1,"child":2})

corr=titanic_dataset[['age','survived','Gender_num']].corr()
sns.heatmap(corr,annot=True,cmap="coolwarm")
#Significance relaationship between


```
