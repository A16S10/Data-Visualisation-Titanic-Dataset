```python
# Defining function to create dictionary with word and its count
def CountFrequncy(my_list):
    frequncy={}
    for word in my_list:
        if word in frequncy:
            frequncy[word]+=1
        else:
            frequncy[word]=1
    for key,value in frequncy.items():
        if value> 50:
            print("%s : % d" % (key, value))
    return frequncy
```

```python
# Creating Dataframe
Top_30_most_used_words=pd.DataFrame(list(Final_dict.items()),columns=['Word','Count'])
# Converting all words in lower case
Top_30_most_used_words['Word']=Top_30_most_used_words['Word'].str.lower()
# Removing Duplicates
Top_30_most_used_words.drop_duplicates()
# Grouping words and taking total count of them
Top_30_most_used_words_new=Top_30_most_used_words.groupby('Word').sum().reset_index()
#Arranging words in descending order and taking top 30
Top_30_most_used_words_new=Top_30_most_used_words_new.sort_values('Count',ascending=False)[:30]
Top_30_most_used_words_new
print(Top_30_most_used_words_new)
```
